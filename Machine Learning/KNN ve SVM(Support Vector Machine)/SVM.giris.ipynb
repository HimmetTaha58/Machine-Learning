### SVM Model

```python
X = df.drop("Turşu", axis=1)
y = df["Turşu"]
```

```python
# SVM support vector machine SVC classifer
from sklearn.svm import SVC
```

```python
# Daha iyi görebilmek için çocuğun yazdığı bir kısım
from svm_margin_plot import plot_svm_boundary
```

### Lineer Model

```python
svc_model = SVC(kernel="linear")
```

```python
svc_model.fit(X, y)
```

## C = 1

```python
svc_model.C
```

```python
plot_svm_boundary(svc_model, X, y)
```

 Bizim burada 2 tane dikkat edilmesi gereken noktamız var svm mantığında model bu eğriyi bu ikisi önemli olduğu için bunların arasına koymaya çalışıyor

```python
svc_model  = SVC(kernel="linear", C=800)
svc_model.fit(X,y)
```

```python
plot_svm_boundary(svc_model, X, y)
```

```python
# C aradaki muallak bölgenin genişliğini daraltıyor yükselince
```

```python
svc_model = SVC(kernel="linear", C = 0.01)
svc_model.fit(X, y)
```

```python
plot_svm_boundary(svc_model, X, y)
```

Burada dikkat edilmesi gereken fazla sayıda veri var ve bunların ortasına oturtmak istiyor o eğriyi o yüzden eğri değişti

C küçükldükçe veriye daha çok bakıyoruz

```python
svc_model = SVC(kernel="rbf")
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

Burada en uç noktadakilere göre sınırlıyor ve kümelenmiş yerlerin dışında kalan noktalara göre eğriyi çiziyor

```python
svc_model = SVC(kernel="sigmoid")
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

Sigmoidin S sini oturtamadığımız için model çuvalladı 

#!# Bunların hiçbirini ezberlemek durumunda değiliz not al kalitelice kenardan bakılabilir

```python
svc_model = SVC(kernel="rbf", gamma=0.003)
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

Gamma ne kadar büyürse bu dışarıda kalan verilere o kadar hassas tepki veriyor

```python
svc_model = SVC(kernel="rbf", gamma=0.3)
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

 Çok sağlam overfitting var yukarıda 

```python
svc_model = SVC(kernel="poly", degree=2)
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

2. Dereceden Polinom oturttu

```python
svc_model = SVC(kernel="poly", degree=12)
svc_model.fit(X, y)
plot_svm_boundary(svc_model, X, y)
```

Bu modelin kullanıcı dostluğu şurada çok kafam ağrımasın aşırı araştırmayayaım kendisi classlasın diyorsan "rbf" kernelini gir başın ağrımaz

## Kendimiz bi oluşturalım

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()

# Burada 10.0 şeklinde yazmazsak hata veriyor integer sayılar negative kuvvetlere izin vermiyor diye

C = 10.0 ** np.arange(-4, 3) # 10 üzeri -4 ten 10 üzeri 2 e kadar diz dedik 3 dahil değil

parameters = {
    "C" : C,
    "kernel" : ["rbf", "linear"]
    }
```

```python
grid = GridSearchCV(svm, parameters)
```

```python
grid.fit(X, y)
```

```python
grid.best_score_
```

```python
1 - grid.best_score_
# 1000 de 2.5 tane yanlış biliyor gayet güzel
```
